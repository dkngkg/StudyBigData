```python
import pandas as pd
import folium
```

```python
h_geoData = pd.read_csv('C:/localRepository/StudyBigData/bus/data/31bus_num_lat_long.csv')  
h_geoData.head()

#, encoding = 'utf-8'
```

```python
# columns 명 변경
h_geoData.rename(columns = {'NAME,C,254': 'station', 'LNG,N,19,15':'lng', 'LAT,N,19,15':'lat'}, inplace=True)
```

```python
h_geoData
```

```python
### folium 사용법
```

```python
map = folium.Map(location= [35.205960, 129.077793], zoom_start=12, tiles ='stamen terrain')
map         
```

```python
### MARKER 표시
```

```python

map = folium.Map(location= [35.205960, 129.077793], zoom_start=15)

folium.Marker(location= [35.205960, 129.077793], popup='동래역',
icon=folium.Icon(color='yellow', icon='plane')).add_to(map)

map 

#icon='plane' , 'bed','home'
```

```python
for station in h_geoData.iterrows():
    print(station)
```

```python
for a, b in h_geoData.iterrows():
    folium.Marker(location=[b['lat'], b['lng']], popup= b['station'], 
    icon=folium.Icon(color='purple', icon='plane')).add_to(map)
  
# 알기) iterrows 를 사용하여 반복문을 사용하면, 2가지 값이 출력된다.  여기선 i 와 store 값이다.
```
